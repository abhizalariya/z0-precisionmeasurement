```python
#putacut by applying range to the particle from the histogram
#Here we try to get maximum particle count by applying diffrent limits to the variables
def particlecuts(N,Psum,E,HE):
    hadrons=[]
    electrons=[]
    tau=[]
    muon=[]
    
    hadrons.append((np.where(N>=15,5,0))) 
    hadrons.append((np.where(N>=5,3,0)))
    hadrons.append((np.where(N>=9,20,0)))
    hadrons.append(np.where(Psum<=90,5,0)) 
    hadrons.append((np.where((E<=60) & (E>=45),1,0)))
    hadrons.append((np.where((E<=90) & (E>=31),2,0)))
    hadrons.append(np.where(HE>=10,1,0)) 
    
    electrons.append(np.where(HE<=20,1,0)) 
    electrons.append(np.where((HE>=0.2)&(HE<=5),0.75,0))
    electrons.append(np.where(N<=2,1,0)) 
    electrons.append(np.where(N<=1.5,2,0)) 
    electrons.append(np.where(Psum<=90,5,0)) 
    electrons.append(np.where(E>=75,2,0)) 
    electrons.append(np.where(E>=70,0.5,0)) 
    electrons.append(np.where(E>=90,5,0)) 
    
    muon.append(np.where(HE<=22,1,0))
    muon.append(np.where(E<=22,1,0)) 
    muon.append(np.where(E<=9,1.5,0))
    muon.append(np.where((Psum>=75),2,0))
    muon.append(np.where( (Psum>=90),2,0)) 
    muon.append(np.where((N<=4) &(N>=1.5),1.5,0))
    
    tau.append(np.where(HE<=25,1,0))
    tau.append(np.where((N<=5) &(N>=1),1,0))   
    tau.append(np.where(Psum <=80,1,0)) 
    tau.append(np.where(Psum <=85,0.25,0)) 
    tau.append(np.where(Psum <=75,1,0)) 
    tau.append(np.where(Psum <=80,2,0)) 
    tau.append(np.where(E<=75,0.5,0)) 
    tau.append(np.where((E<=30) &(E>=5),1,0)) 
    
    hadrons= np.sum(hadrons, axis=0)
    electrons= np.sum(electrons, axis=0)
    tau= np.sum(tau, axis=0)
    muon= np.sum(muon, axis=0)
    
    maximum=np.amax([[hadrons],[electrons],[tau],[muon]], axis=0)

    hadron=np.sum(np.where(hadrons>=maximum, 1,0))
    electron=np.sum(np.where(electrons>=maximum, 1,0))
    tau=np.sum(np.where(tau>=maximum, 1,0))
    muon=np.sum(np.where(muon>=maximum, 1,0))
    
    return np.array([hadron, electron, tau,muon])
```

```python
matrix=[]
error_matrix=[]
criteria=[]


for i in ['qq.root','ee.root', 'tt.root','mm.root']:
    file = uproot.open(path_data+i)
    ttree_name = 'myTTree'
    file[ttree_name].keys()
    branches = file[ttree_name].arrays()
    Pcharged = 'Pcharged'
    NCharged='Ncharged'
    E_ecal='E_ecal'
    E_hcal='E_hcal'
    cos_thet='cos_thet'
    E_lep='E_lep'
    

    pchar = ak.to_numpy(branches[Pcharged]) # See Docu (https://awkward-array.org/how-to-convert-numpy.html) for more conversions
    Nchar = ak.to_numpy(branches[NCharged])
    E_ecal = ak.to_numpy(branches[E_ecal])
    E_hcal = ak.to_numpy(branches[E_hcal])
    cos_thet =ak.to_numpy(branches[cos_thet])
    E_lep = ak.to_numpy(branches[E_lep])
    
    criteria.append([pchar, Nchar, E_ecal, E_hcal, cos_thet,E_lep]) 
    listen=np.array(np.full(4,0))
    countparticles=particlecuts(Nchar,pchar,E_ecal,E_hcal)
    matrix.append(countparticles/len(Nchar))
    error_matrix.append(np.sqrt(countparticles/len(Nchar)**2+countparticles**2/len(Nchar)**3))
    


matrix=np.array(matrix)
error_matrix=np.array(error_matrix)
print('Efficiency matrix')
print(matrix)
print('Error efficiency matrix')
print(error_matrix)
```

```python
def particlecuts2(N,Psum,E,HE):
    hadrons=[]
    electrons=[]
    tau=[]
    muon=[]
    
    hadrons.append((np.where(N>=15,5,0))) 
    hadrons.append(np.where(Psum<=90,5,0)) 
    hadrons.append((np.where((E<=90) & (E>=31),2,0)))
    hadrons.append(np.where(HE>=10,1,0)) 
    
     
    electrons.append(np.where(N<=2,1,0))  
    electrons.append(np.where(Psum<=90,5,0))   
    electrons.append(np.where(E>=90,5,0))
    electrons.append(np.where(HE<=20,1,0))
    
    muon.append(np.where((N<=3),2,0))
    muon.append(np.where((Psum>=20),2,0))
    muon.append(np.where(E<=22,1,0))
    muon.append(np.where(HE<=14,1,0))
     
   
    
    
    tau.append(np.where((N<=5) &(N>=1),1,0))    
    tau.append(np.where(Psum <=80,2,0)) 
    tau.append(np.where(E<=75,0.5,0)) 
    tau.append(np.where(HE<=25,1,0))
     
    
    hadrons= np.sum(hadrons, axis=0)
    electrons= np.sum(electrons, axis=0)
    tau= np.sum(tau, axis=0)
    muon= np.sum(muon, axis=0)
    
    maximum=np.amax([[hadrons],[electrons],[tau],[muon]], axis=0)

    hadron=np.sum(np.where(hadrons>=maximum, 1,0))
    electron=np.sum(np.where(electrons>=maximum, 1,0))
    tau=np.sum(np.where(tau>=maximum, 1,0))
    muon=np.sum(np.where(muon>=maximum, 1,0))
    
    return np.array([hadron, electron, tau,muon])

def array(ctrk,sump,ecal,hcal):
    
    result = []
    for i in range(0,len(ctrk)):
        result.append(particlecuts2(ctrk[i],sump[i],ecal[i],hcal[i]))
    return result
```

```python
def particles(ctrk,sump,ecal,hcal):
    spare = []
    name1 = []
    particle_list = array(ctrk,sump,ecal,hcal)
    for particle in particle_list:
        name = []
        indices = [i for i, x in enumerate(particle) if x == max(particle)]
        if 0 in indices:
            name = 1
        if 1 in indices:
            name = 2
        if 2 in indices:
            name = 3
        if 3 in indices:
            name = 4
        if len(indices) != 1:
            name=0
            spare.append(np.array(indices)+1)
        name1.append(name)
    return [name1,spare]

data_particles,spare = particles(data5_ctrk,data5_sump,data5_ecal,data5_hcal) 
```
